```python
# ------------------------------
# Imports
# ------------------------------
import pandas as pd
import numpy as np
import time

# ------------------------------
# Dataset paths
# ------------------------------
CITY_DATA = {
    'chicago': 'chicago.csv',
    'new york city': 'new_york_city.csv',
    'washington': 'washington.csv'
}

# ------------------------------
# Functions
# ------------------------------

def get_filters():
    """
    Asks the user to specify a city, month, and day to analyze.
    Ensures inputs are valid.
    """
    print("Hello! Let's explore some US bikeshare data!")
    
    # --- CITY ---
    while True:
        city = input("\nChoose a city: Chicago, New York City, Washington\n> ").strip().lower()
        if city in CITY_DATA:
            break
        print("âŒ Invalid city. Try again.")
    
    # --- LOAD CITY DATA FIRST ---
    df_city = pd.read_csv(CITY_DATA[city])
    df_city.columns = df_city.columns.str.strip()
    df_city['Start Time'] = pd.to_datetime(df_city['Start Time'], dayfirst=True, errors='coerce')
    df_city['Month'] = df_city['Start Time'].dt.month_name().str.lower()
    df_city['Weekday'] = df_city['Start Time'].dt.day_name().str.lower()
    
    # Get unique months and weekdays in this city
    available_months = sorted(df_city['Month'].dropna().unique())
    available_days = sorted(df_city['Weekday'].dropna().unique())
    
    # --- MONTH ---
    while True:
        month = input(f"\nAvailable months for {city.title()}: {', '.join([m.title() for m in available_months])}, or 'all'\n> ").strip().lower()
        if month in available_months or month == 'all':
            break
        print("âŒ Invalid month. Try again.")
    
    # --- DAY ---
    while True:
        day = input(f"\nAvailable days for {city.title()}: {', '.join([d.title() for d in available_days])}, or 'all'\n> ").strip().lower()
        if day in available_days or day == 'all':
            break
        print("âŒ Invalid day. Try again.")
    
    print('-'*40)
    print(f"âœ… You selected:\nCity: {city.title()}\nMonth: {month.title()}\nDay: {day.title()}")
    print('-'*40)
    
    return city, month, day


def load_data(city, month, day):
    """Load data and filter by month and day."""
    
    df = pd.read_csv(CITY_DATA[city])
    df.columns = df.columns.str.strip()
    
    # Parse datetime safely
    df['Start Time'] = pd.to_datetime(df['Start Time'], dayfirst=True, errors='coerce')
    df['Month'] = df['Start Time'].dt.month_name().str.lower()
    df['Weekday'] = df['Start Time'].dt.day_name().str.lower()
    
    # Filter month/day
    if month != 'all':
        df = df[df['Month'] == month]
    if day != 'all':
        df = df[df['Weekday'] == day]
    
    # Fill missing columns for Washington or incomplete datasets
    for col in ['User Type', 'Gender', 'Birth Year']:
        if col not in df.columns:
            df[col] = pd.NA
    
    if df.empty:
        print(f"\nâš ï¸ No data found for city={city.title()}, month={month.title()}, day={day.title()}.")
        print("ðŸ‘‰ Try using 'all' for month or day to see available data.\n")
    
    return df


def user_stats(df, city):
    """Display statistics about bikeshare users."""
    start_time = time.time()
    
    if df.empty:
        print(f"\nâš ï¸ No user statistics available for city={city.title()} (empty dataset).")
        print('-'*40)
        return
    
    print('\nCalculating User Stats...\n')
    
    # --- User Types ---
    usertypes = df['User Type'].value_counts()
    print(f"Subscribers: {usertypes.get('Subscriber', 0)}")
    print(f"Customers: {usertypes.get('Customer', 0)}")
    
    # --- Gender ---
    if df['Gender'].notna().any():
        print("\nGender distribution:")
        print(df['Gender'].value_counts())
    else:
        print("\nNo gender data available for this city.")
    
    # --- Birth Year ---
    if df['Birth Year'].notna().any():
        birth_years = df['Birth Year'].dropna()
        print(f"\nEarliest birth year: {int(birth_years.min())}")
        print(f"Most recent birth year: {int(birth_years.max())}")
        print(f"Most common birth year: {int(birth_years.mode()[0])}")
    else:
        print("\nNo birth year data available for this city.")
    
    print(f"\nThis took {time.time() - start_time:.4f} seconds.")
    print('-'*40)


def raw_data(df):
    """Displays 5 lines of raw data at a time."""
    if df.empty:
        print("\nâš ï¸ No raw data available for the selected filters.")
        return
    
    i = 0
    while True:
        show = input("\nWould you like to see 5 lines of raw data? Enter yes or no.\n> ").strip().lower()
        if show != 'yes':
            break
        print(df.iloc[i:i+5])
        i += 5
        if i >= len(df):
            print("\nâœ… End of data reached.")
            break


# ------------------------------
# Main Program
# ------------------------------
def main():
    while True:
        city, month, day = get_filters()
        df = load_data(city, month, day)
        
        # Only show stats and raw data if DataFrame has data
        if not df.empty:
            user_stats(df, city)
            raw_data(df)
        
        restart = input("\nWould you like to restart? Enter yes or no.\n> ").strip().lower()
        if restart != 'yes':
            print("\nGoodbye!")
            break


if __name__ == "__main__":
    main()
```
